```python
import numpy as np
import seaborn as sns
from pca import pca, standardizeData
import pandas as pd
    
# test_Data = np.random.randint(1, 50, (5,10))
# print(test_Data)
rawData = pd.read_csv('../../health_data.csv')

test_data = standardizeData(rawData.to_numpy())
simplified_data = standardizeData(rawData.drop(['AGE','SEX','BP','S1','S3','S4'], axis=1).to_numpy())
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# scaler = StandardScaler(copy=True)
# scaler.fit(test_Data)
# scaledData = scaler.transform(test_Data)
scaledData = standardizeData(simplified_data)
pca_util = PCA(n_components=2)
result = pca_util.fit_transform(scaledData)
print(result)
df = pd.DataFrame(data=result)
ax = sns.scatterplot(data=df, x=0, y=1)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
```

```python

values, vectors, variance, projectedData, covMatrix = pca(2, simplified_data)

# print(values)
# print(vectors)
# print(variance)
print(projectedData)

df = pd.DataFrame(data=projectedData)
ax = sns.scatterplot(data=df, x=0, y=1)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
```
